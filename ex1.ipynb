```python
import matplotlib.pyplot as plt

from estatistica import coef_correlacao, covariancia, desvio_padrao, media, qui_quadrado, pares
from pseudoaleatorio import mcl
```

```python
def run_plot(a, c, M, seed=10, it=10_000):
    print("a = {}, M = {}, c = {}".format(a, M, c))
    print("seed = {}, {} iterações".format(seed, it))

    x = seed
    amostra = [x / M]

    for _ in range(it):
        x = mcl(a, x, c, M)
        amostra.append(x / M)

    print("\nmédia = {:.5f}".format(media(amostra)))
    print("desvio padrão = {:.5f}".format(desvio_padrao(amostra)))
    print("covariância = {:.5f}".format(covariancia(amostra)))
    print("coeficiente de correlação = {:.5f}".format(coef_correlacao(amostra)))
    print("qui-quadrado = {:.5f}".format(qui_quadrado(amostra)))
    
    x, y = pares(amostra)
    plt.scatter(x, y, s=1)
    plt.title("a = {}, c = {}, M = {}\nseed = {}, {} iterações".format(a, c, M, seed, it))
    plt.axis('square')
    fig = plt.gcf()
    fig.set_size_inches(6, 6)
    plt.show()
```

```python
run_plot(45, 1, 1024)
```

```python
run_plot(45, 0, 1024)
```

```python
run_plot(383, 263, 1000)
```

```python
run_plot(97, 0, 2**17)
```

```python
import matplotlib.pyplot as plt
from estatistica import coef_correlacao, covariancia, desvio_padrao, media
from pseudoaleatorio import gerar, uniforme, exponencial, weibull, normal, erlang, poisson, geometrica
```

```python
def show_stats(amostra, dist):
    print("distribução", dist)
    print("\nmédia = {:.5f}".format(media(amostra)))
    print("desvio padrão = {:.5f}".format(desvio_padrao(amostra)))
    # print("covariância = {:.5f}".format(covariancia(amostra)))
    # print("coeficiente de correlação = {:.5f}".format(coef_correlacao(amostra)))
    
    # fig, axs = plt.subplots(1, 2)
    # fig.set_size_inches(12, 6)
    
    # axs[0].hist(amostra, bins=30, density=True)
    # axs[0].set_title("Distribuição {}".format(dist))
    
    # x, y = pares(amostra)
    # axs[1].scatter(x, y, s=.5)
    # axs[1].axis('square')
    
    # plt.show()
    
    plt.title(dist)
    plt.hist(amostra, bins=50, density=True)
    fig = plt.gcf()
    fig.set_size_inches(8, 6)
    plt.show()
```

```python
n = 10_000
seed = 42
```

```python
start = 3
end = 6

amostra = gerar(uniforme(start, end, seed), n)
show_stats(amostra, "Uniforme\n[{:.2f}, {:.2f}]".format(start, end))
```

```python
alpha = 3

amostra = gerar(exponencial(alpha, seed), n)
show_stats(amostra, "Exponencial\nalpha = {:.5f}".format(alpha))
```

```python
alpha = 3
beta = 1

amostra = gerar(weibull(alpha, beta, seed), n)
# show_stats(amostra, "Weibull\nalpha={:.2f}, beta={:.2f}".format(alpha, beta))
```

```python
mu = 5
sigma = 3

amostra = gerar(normal(mu, sigma, seed), n)
show_stats(amostra, "Normal\nmu={:.2f}, sigma={:.2f}".format(mu, sigma))
```

```python
k = 2
alpha = 3

amostra = gerar(erlang(k, alpha, seed), n)
show_stats(amostra, "Erlang\nk={:.2f}, alpha={:.2f}".format(k, alpha))
```

```python
lamb = 5

amostra = gerar(poisson(lamb, seed), n)
show_stats(amostra, "Poisson\nlambda={:.2f}".format(lamb))
```

```python
p = 0.5

amostra = gerar(geometrica(p, seed), n)
show_stats(amostra, "Geometrica\np={:.2f}".format(p))
```
